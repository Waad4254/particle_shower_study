# 1. Embedding via SVD.
In this exercise, we would like to use the SVD to build a 50-dimensional latent space representation of MNIST data set, i.e. reduce its dimensionality from 784 (28 × 28) to 50.

```python
# read MNIST training data
df = pd.read_csv('./mnist_train.csv', dtype=np.dtype(np.float32) )
X = df.iloc[:, 1:].to_numpy() / 255.0       # values are scaled to be between 0 and 1
y = df.iloc[:, 0].to_numpy()                # labels of images
XT = X.transpose()
```

• generate the truncated SVD of the (transposed) data matrix

```python
U, s, VT = svd(XT,full_matrices=False)
```

```python
Sigma = np.diag(s)
```

```python
k = 50
```

```python
U=U[:,:k]
Sigma=Sigma[:k,:k]
VT=VT[:k,:]
```

```python
X_truncated = U @ Sigma @ VT
```

• test the quality of the representation by reconstructing a few randomly chosen examples

```python
fig, ax = plt.subplots(5, 2, figsize=(10, 40))
curr_fig = 0
for x in [1,6,9,20,100]:
    imag = XT[:, x].reshape((28,28))
    image_trunc = X_truncated[:, x].reshape((28,28))
    ax[curr_fig][0].imshow(image_trunc, cmap='gray')
    ax[curr_fig][0].set_title("k = " + str(k))
    ax[curr_fig, 0].axis('off')
    ax[curr_fig][1].set_title("Original Image")
    ax[curr_fig][1].imshow(imag,cmap='gray')
    ax[curr_fig, 1].axis('off')
    curr_fig += 1
plt.show()
```

Very briefly, answer the following questions:
• Suppose we wanted to build a classifier for the MNIST data. Based on your observations above,
is it possible to use the latent space representation as input to the classifier? Any advantages in
doing so?

Yes it's possible. The classification will be faster and more reliable when working in the latent space than working in pixel space. 

• What is the cost of performing a randomized SVD on a matrix of size M × N to generate a low
rank approximation of rank r? How does this compare with the cost of performing the full SVD
on the matrix and then truncating it.

The cost of full SVD on the matrix X is $O(MN^2) + O(N^3)$ whereas the cost of the randomized SVD is $O(NL^2) + O(L^3)$, where L << N, M is the number of examples, N is the feature dimension, L is the reduced feature dimension. 

This lower dimension L is obtained by first constructing a matrix $Q$ whose columns form an orthonormal basis for the range of X. This is done by drawing L Gaussian random vectors and projecting them using the linear map X(sampling the data matrix $XΩ$) followed by a $QR$ factorization. Finally, compute the SVD on the small matrix $B = Q^T A$. Performing the SVD on this smaller matrix is much cheaper than performing it on the original one.

```python
# read MNIST training data
df = pd.read_csv('./mnist_train.csv')
X = df.iloc[:, 1:].to_numpy(dtype=np.float32) / 255.0   # values scaled to [0, 1]
# X = X.reshape((X.shape[0], 28, 28))

batch_size = 256
batches = X.shape[0]//batch_size + 1
X_batched = np.array_split(X, [i*batch_size for i in range(1,batches)])
```

## 2.1 Network

```python
# define network
class MLP(nn.Module):
    layers_sizes: List[int]
    
    @nn.compact
    def __call__(self, x):
        # Encoder
        for l in range(0, len(self.layers_sizes)):
            x = nn.Dense(
                self.layers_sizes[l], name=f"Encoder_Hidden_layer_{l}"
            )(x)
            x = nn.selu(x)
        # Decoder
        for l in range(len(self.layers_sizes)-2, -1, -1):
            x = nn.Dense(
                self.layers_sizes[l], name=f"Decoder_Hidden_layer_{l}"
            )(x)
            x = nn.selu(x)
        x = nn.Dense(784, name="Output_layer")(x)
        # x = nn.sigmoid(x)
        return x

mlp = MLP(layers_sizes=[100, 30])
seed = 215
rkey = jax.random.PRNGKey(seed)
mlp_params = mlp.init(rkey, X)
```

```python
list(range(5-2, -1, -1))
```

```python
print(jax.tree_util.tree_map(lambda x: x.shape, mlp_params))   # print params shapes
print(mlp.tabulate(jax.random.PRNGKey(0), X[0:1]))            # model details
```

```python
X[:4]
```

```python
nn.sigmoid(mlp.apply(mlp_params, X[:4])).shape
```

## 2.2 Loss

```python
# define loss
@jax.jit
def loss(mlp_params, x_batched):
    return jnp.mean(optax.sigmoid_binary_cross_entropy(mlp.apply(mlp_params, x_batched), x_batched),)
```

```python
loss(mlp_params, X[:4])
```

## 2.3 Train

```python
# training step
@jax.jit
def train_step(mlp_params, opt_state, x_batch):
    # compute the gradient for this batch
    grads = jax.grad(loss)(mlp_params, x_batch)
    # compute the parameters (and state) updates
    updates, opt_state = opt.update(grads, opt_state)
    # apply the updates to the network parameters
    params = optax.apply_updates(mlp_params, updates)
    return params, opt_state
```

```python
# training loop
learning_rate = 1
momentum = 0.9
num_epochs = 5

opt = optax.sgd(learning_rate, momentum)
opt_state = opt.init(mlp_params)

for epoch in range(num_epochs):
    for b in range(batches):
        mlp_params, opt_state = train_step(mlp_params, opt_state, X_batched[b]) 
        if b % 100 == 0:
            print(f'Loss step {b}: ', loss(mlp_params, X_batched[b]))
```

## 2.4 Test

```python
plt.figure()
for i in range(5):
    plt.subplot(2, 5, i+1, xticks=[], yticks=[])
    k = np.random.randint(X.shape[0])
    image = X[k].reshape(28, 28)
    plt.imshow(image, cmap='gray')
    plt.subplot(2, 5, i+6, xticks=[], yticks=[])
    rec_image = jax.nn.sigmoid(mlp.apply(mlp_params, X[k]).squeeze()).reshape(28, 28)
    plt.imshow(rec_image, cmap='gray')
```

```python
# read MNIST training data
df = pd.read_csv('./mnist_train.csv')
X = df.iloc[:, 1:].to_numpy(dtype=np.float32) / 255.0   # values scaled to [0, 1]
X = X.reshape((X.shape[0], 28, 28, 1))

batch_size = 256
batches = X.shape[0]//batch_size + 1
X_batched = np.array_split(X, [i*batch_size for i in range(1,batches)])
```

## 3.1 Network

```python
7*7*64
```

```python
rkey
```

```python
dummy_input = np.zeros((2, 3, 3, 64))
conv = nn.ConvTranspose(32, (3, 3), strides=(2,2), padding="VALID")
conv_params = conv.init(rkey, dummy_input)
dummy_input = conv.apply(conv_params, dummy_input)
print(dummy_input.shape)

conv = nn.ConvTranspose(16, (3, 3), strides=(2,2), padding="SAME")
conv_params = conv.init(rkey, dummy_input)
dummy_input = conv.apply(conv_params, dummy_input)
print(dummy_input.shape)

conv = nn.ConvTranspose(1, (3, 3), strides=(2,2), padding="SAME")
conv_params = conv.init(rkey, dummy_input)
dummy_input = conv.apply(conv_params, dummy_input)
print(dummy_input.shape)
```

```python
# define network
class CNN(nn.Module):    
    @nn.compact
    def __call__(self, x):
        # Encoder
        x = nn.Conv(16, (3, 3), padding="SAME")(x)
        x = nn.selu(x)
        x = nn.max_pool(x, (2, 2), strides=(2,2))
        
        x = nn.Conv(32, (3, 3), padding="SAME")(x)
        x = nn.selu(x)
        x = nn.max_pool(x, (2, 2), strides=(2,2))
        
        x = nn.Conv(64, (3, 3), padding="SAME")(x)
        x = nn.selu(x)
        x = nn.max_pool(x, (2, 2), strides=(2,2))
        # Decoder
        x = nn.ConvTranspose(32, (3, 3), strides=(2,2), padding="VALID")(x)
        x = nn.selu(x)
        x = nn.ConvTranspose(16, (3, 3), strides=(2,2), padding="SAME")(x)
        x = nn.selu(x)
        x = nn.ConvTranspose(1, (3, 3), strides=(2,2), padding="SAME")(x)
        return x

model = CNN()
seed = 215
rkey = jax.random.PRNGKey(seed)
params = model.init(rkey, X)
```

```python
print(jax.tree_util.tree_map(lambda x: x.shape, params))   # print params shapes
print(model.tabulate(jax.random.PRNGKey(0), X[0:2]))            # model details
```

```python
model.apply(params, X[:4]).shape
```

```python
nn.sigmoid(model.apply(params, X[:4])).shape
```

## 2.2 Loss

```python
# define loss
@jax.jit
def loss(params, x_batched):
    return jnp.mean(optax.sigmoid_binary_cross_entropy(model.apply(params, x_batched), x_batched),)
```

```python
loss(params, X[:4])
```

## 2.3 Train

```python
# training step
@jax.jit
def train_step(params, opt_state, x_batch):
    # compute the gradient for this batch
    grads = jax.grad(loss)(params, x_batch)
    # compute the parameters (and state) updates
    updates, opt_state = opt.update(grads, opt_state)
    # apply the updates to the network parameters
    params = optax.apply_updates(params, updates)
    return params, opt_state
```

```python
# training loop
learning_rate = 0.1
momentum = 0.9
num_epochs = 5

opt = optax.sgd(learning_rate, momentum)
opt_state = opt.init(params)

for epoch in range(num_epochs):
    for b in range(batches):
        params, opt_state = train_step(params, opt_state, X_batched[b]) 
        if b % 100 == 0:
            print(f'Loss step {b}: ', loss(params, X_batched[b]))
```

## 2.4 Test

```python
plt.figure()
for i in range(5):
    plt.subplot(4, 5, i+1, xticks=[], yticks=[])

    k = np.random.randint(X.shape[0])
    image = X[k]
    plt.imshow(image, cmap='gray')
    
    plt.subplot(4, 5, i+6, xticks=[], yticks=[])
    rec_image_cnn = jax.nn.sigmoid(model.apply(params, X[k]).squeeze())
    plt.imshow(rec_image_cnn, cmap='gray')
    
    plt.subplot(4, 5, i+11, xticks=[], yticks=[])
    rec_image = jax.nn.sigmoid(mlp.apply(mlp_params, X[k].reshape(1, -1)).squeeze()).reshape(28, 28)
    plt.imshow(rec_image, cmap='gray')
    
    plt.subplot(4, 5, i+16, xticks=[], yticks=[])
    image_trunc = X_truncated[:, k].reshape((28,28))
    plt.imshow(image_trunc, cmap='gray')
```
