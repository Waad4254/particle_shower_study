We revisit the clustering problem of N unlabeled data and examine it from a statistical perspective.
Given N points with D dimensional features in R, we may fit K D-dimensional Gaussians to the data
set in order to assign the highest probability to the training data (maximize the likelihood). In this
formulation, the inclusion of point n in cluster k is no longer a binary choice, but is defined by a
vector rnk Σ that reflects the responsibility (membership probability) of cluster k for data point n, with K
k=1 rnk = 1

• list one or mores advantages that such a clustering method may have over the basic K-means
approach we saw earlier

**Can be used for non-spherical clusters**

**Can be used to generate new data**

• write the objective function to be optimized to find the clustering of the data points and compare
it with the objective used for the simple K-means approach

GMM loss function: $$ \sum \limits _{n=1}^N \sum \limits _{k=1}^k r_{nk} [log  π_k + log N(x_n | μ_k, Σ_k)] $$

K-means loss function: $$\sum \limits _{n=1}^N \sum \limits _{k=1}^K r_{nk} ||x_n - μ_k||^2  $$

The loss function in k-means minimizes the sum of squared distance while the loss function in GMM minimizes the negative log likelihood 

• briefly explain how your earlier clustering algorithm will change (section 8.7.3 of Reference 1
describes the algorithm)

First: The loss function will change. 
Second: The outer loop with do the **E step**, which computes the responsibility of cluster k for generating data point n using the current parameters µ and Σ. 

Third: The inner loop will do the **M step**, which maximizes the log of likelihood.

Forth: the updated µ and Σ resuilting from the inner loop will be used to recalculate the responsibilities. 

We keep doing that until we converge.
