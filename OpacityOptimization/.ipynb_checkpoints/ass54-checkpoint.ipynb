```python
import jax
import jax.numpy as jnp 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from jax import random
import optax
import math
from tqdm import tqdm
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('iris.csv')
```

```python
# make the labels 0,1 and 2
df['species'].replace(["Iris-setosa","Iris-versicolor","Iris-virginica"], [0,1,2], inplace=True)
```

```python
# generate X and Y matrices, adding the ``1'' feature for the bias
N, D = df.shape
X = df.iloc[:, 0:D-1].to_numpy().astype(np.float32)
X = np.concatenate((X, np.ones((N,1))), axis=1)
Y = df.iloc[:, D-1].to_numpy().astype(np.float32)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y)
```

```python
@jax.jit
def loss_per_example(W, x, y):
    return jnp.sum(y*jax.nn.log_softmax(x@W))
```

```python
# loss function for a batch
@jax.jit
def loss(W, X, Y):
    totalLoss = 0
    for x, y in zip(X, Y):
        totalLoss += loss_per_example(W, x, y)
    return -totalLoss/len(Y)
```

```python
def gradient_NLL(W, X, Y):
    return -X.T@(Y-jax.nn.softmax(X@W))/len(Y)
```

```python
onehot_encoder = OneHotEncoder(sparse=False)
Y_train_onehot = onehot_encoder.fit_transform(y_train.reshape(-1,1))
W = np.random.normal(size=(X.shape[1], Y_train_onehot.shape[1]))
```

```python
init_learning_rate = 0.1
optimizer = optax.inject_hyperparams(optax.rmsprop)(learning_rate=init_learning_rate)
state = optimizer.init(W)
```

```python
lossArray = []
thetaarray = []
B = 10
```

```python
#np.split(train, B)
```

```python
for i in tqdm(range(2000)):
    for batch_indices in np.split(np.random.permutation(np.arange(0, len(X_train))), B):
        batch_indices = batch_indices.astype(int)
        value, grad = jax.value_and_grad(loss)(W, X_train[batch_indices], Y_train_onehot[batch_indices])
        updates, state = optimizer.update(grad, state)
        W = optax.apply_updates(W, updates)
        lossArray.append(value)
        thetaarray.append(W)
```

```python
fig, (ax1) = plt.subplots(nrows=1, squeeze=True)
fig.suptitle('loss value\ k iteration')
plt.xlabel('K iterations')

ax1.semilogy(range(len(lossArray)), lossArray)
ax1.set(ylabel='Training loss')
ax1.label_outer()

plt.show()
```

```python
print(W)
```

```python
predicted = X_test@W 
predicted_y = np.zeros(y_test.shape)
```

```python
for x in range(predicted.shape[0]):
    index = np.argmax(predicted[x])
    predicted_y[x]=index
```

```python
print(predicted_y)
print(y_test)
```

```python
import seaborn as sns
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, predicted_y)
```

```python
sns.heatmap(confusion_matrix, cmap="Blues", annot=True, fmt="g", 
            xticklabels=['Setosa','Versicolor','Virginica'],
            yticklabels=['Setosa','Versicolor','Virginica'])
```

The confusion matrix shows the performance of the classifier in which all iris types were classified correctly except for one virginica that was classified as versicolor

```python
gradient_NLL(W,X_train, Y_train_onehot)
```

```python
jax.grad(loss)(W, X_train, Y_train_onehot)
```

As can be seen above the direct gradient computation gives the same result as the
jax-computed gradient.

The mathematical expression for the regularization term (L1 norm):

$$ NLL(W) = \frac{-1}{N} \sum \limits _{n=1}^N \sum \limits _{c=1}^C y_{nc} log(μ_{nc}) + λ|W| $$

This will force some of the features' coefficients to be so small when the parameter λ is large. In simple words, it encourages sparsity.
